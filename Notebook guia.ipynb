## Importación de bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Creación de datos

```python
m = 0.8
b = 1.0
n = 171
x = np.random.uniform(0, 10, n)
y = m * x + b + np.random.normal(0, 2, n) 
```

## Graficación de los datos

```python
plt.figure(figsize = (8, 5))
plt.scatter(x, y, s = 2.1)
plt.show()
```

## Regresión lineal usando polyfit

```python
recta_parametros = np.polyfit(x, y, 1)
recta_parametros
```

## Construcción de la recta

```python
m_regresion = recta_parametros[0]
b_regresion = recta_parametros[1]
y_regresion = m_regresion * x + b_regresion
```

## Gráficación de los datos junto con la recta obtenida

```python
plt.figure(figsize = (8, 5))
plt.scatter(x, y, s = 2.1, label = "datos")
plt.plot(x, y_regresion, color = "black", label = "regresión")
plt.legend(loc = "best")
plt.show()
```
